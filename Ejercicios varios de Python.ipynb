```python
def IMC_input_print():
  peso=float(input('introduce tu peso en kg'))
  altura=float(input('introduce tu altura en metros'))
  peso=float (peso)
  altura=float(altura)
  print(f"tu IMC es {peso / (altura ** 2)}")
IMC_input_print()
```
