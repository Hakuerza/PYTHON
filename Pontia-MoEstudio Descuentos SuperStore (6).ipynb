<span style="font-size: 15pt; font-weight: bold; color: #0098cd;">1.3. Generación de variables calculadas.</span>

Genere a partir del dataframe cargado las siguientes variables calculadas.

- Sales/Quantity: Precio de Venta por unidad de producto en cada operación de venta
- Profit/Quantity: Beneficio por unidad de producto en cada operación de venta
- flag_descuento: Marca con "Con Descuento" si esa operacion de venta ha tenido descuento (esto es descuento>1), "Sin Descuento" si no.
- Year: Variable con el año de la fecha de la operación de venta.

Respuesta:

```python
#Sales/Quantity: Precio de Venta por unidad de producto en cada operación de venta
df['Sales/Quantity'] = df['Sales']/df['Quantity']
df.head()

#Profit/Quantity: Beneficio por unidad de producto en cada operación de venta
df['Sales/Quantity'] = df['Sales']/df['Quantity']
df.head()

#flag_descuento: Marca con "Con Descuento" si esa operacion de venta ha tenido descuento (esto es descuento>1), "Sin Descuento" si no.
df['flag_descuento'] = np.where(df['Discount']>1, 'Con Descuento', 'Sin Descuento')
df.head()

#Year: Variable con el año de la fecha de la operación de venta.
df['Year'] = pd.to_datetime(df['Order Date']).dt.year
df.head()
```
