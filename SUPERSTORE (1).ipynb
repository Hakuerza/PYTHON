<span style="font-size: 15pt; font-weight: bold; color: #0098cd;">1.3. Generación de variables calculadas.</span>

Genere a partir del dataframe cargado las siguientes variables calculadas.

- Sales/Quantity: Precio de Venta por unidad de producto en cada operación de venta
- Profit/Quantity: Beneficio por unidad de producto en cada operación de venta
- flag_descuento: Marca con "Con Descuento" si esa operacion de venta ha tenido descuento (esto es descuento>1), "Sin Descuento" si no.
- Year: Variable con el año de la fecha de la operación de venta.

Respuesta:

```python
# 1. Sales/Quantity: Precio de Venta por unidad de producto en cada operación de venta
df['Sales quantity'] = df['Sales'] / df['Quantity']
```

```python
df.head()
```

```python
# 2. Profit/Quantity: Beneficio por unidad de producto en cada operación de venta
df['unit profit'] = df['Profit'] / df['Quantity']
```

```python
df.head ()
```

```python
# 3. flag_descuento: Marca con "Con Descuento" si esa operacion de venta ha tenido descuento (esto es descuento>1), "Sin Descuento" si no.

df['flag_descuento'] = df['Discount'].apply(lambda x: 'Con Descuento' if x > 0 else 'Sin Descuento')
```

```python
df.head ()
```

```python
# 4. Año de la fecha (Year)
df['Year'] = pd.to_datetime(df['Order Date']).dt.year

df.head ()
```

<span style="font-size: 15pt; font-weight: bold; color: #0098cd;">2.1. Gráfico 1.</span>

Respuesta:

```python
#comparativa de ventas con y sin descuento
```

```python
con_sin_desc = df.groupby('flag_descuento').agg(
    Count=('Sales', 'count')
).reset_index()
con_sin_desc
```

```python
import plotly.express as px

fig = px.bar(
    con_sin_desc,
    x='flag_descuento',
    y='Count',
    title='Cantidad de Ventas según el Descuento',
    labels={'flag_descuento': 'Categoría de Descuento', 'Count': 'Cantidad de Ventas'},
    color='flag_descuento',
    template = 'simple_white',
)
fig.show()
```

En este gráfico se puede apreciar que la cantidad de ventas con descuentos es de 5196 y sin descuentos es de 4798, con lo cual vemos el el porcentage no es muy significativo.


```python
# La misma de antes pero separando por años
```

```python
con_años = df.groupby(['Year', 'flag_descuento']).agg(
    Count=('Sales', 'count')  # Cantidad de ventas
).reset_index()

con_años
```

```python
fig1 = px.bar(
    con_años,
    x='Year',
    y='Count',
    color='flag_descuento',
    barmode='group',
    title='Cantidad de Ventas según Descuento y Año',
    labels={'flag_descuento': 'Categoría de Descuento', 'Count': 'Cantidad de Ventas', 'Year': 'Año'},
    color_discrete_map={'Con Descuento': 'skyblue', 'Sin Descuento': 'orange'},
    template = 'simple_white',
)

fig1.show()
```

En este gráfico mostramos las ventas con y sin descuento por años.
En el comprobamos que las ventas con descuento es mayor que las ventas sin descuento.
Comprobamos que las ventas han ido en aumento en ambas,
pero se ve  que la empresa tiene muchos productos en oferta y pueda ser el motivo de que las ventas con descuento sean mayores.

Conclusiones: La empresa ha seguido la tendencia de tener una gran parte de su catalogo con descuento aplicado



<span style="font-size: 15pt; font-weight: bold; color: #0098cd;">2.2. Gráfico 2.</span>

Respuesta:

```python
comparativa = df.groupby(['Year', 'flag_descuento']).agg(
    Total_Sales=('Sales', 'sum'),  # Sumar las ventas por año y descuento
    Total_Profit=('Profit', 'sum')  # Sumar los beneficios por año y descuento
).reset_index()

comparativa
```

```python
import plotly.express as px
```

```python
import plotly.graph_objects as go
```

```python
# Make sure to execute the cell containing this code first:
comparativa = df.groupby(['Year', 'flag_descuento']).agg(
    Total_Sales=('Sales', 'sum'),  # Sumar las ventas por año y descuento
    Total_Profit=('Profit', 'sum')  # Sumar los beneficios por año y descuento
).reset_index()

fig = go.Figure()

fig.add_trace(go.Bar(
    x=comparativa[comparativa['flag_descuento'] == 'Con Descuento']['Year'],
    y=comparativa[comparativa['flag_descuento'] == 'Con Descuento']['Total_Sales'],
    name='Ventas con Descuento',
    marker=dict(color='skyblue'),
))


fig.add_trace(go.Bar(
    x=comparativa[comparativa['flag_descuento'] == 'Sin Descuento']['Year'],
    y=comparativa[comparativa['flag_descuento'] == 'Sin Descuento']['Total_Sales'],
    name='Ventas sin Descuento',
    marker=dict(color='orange')
))


fig.add_trace(go.Scatter(
    x=comparativa[comparativa['flag_descuento'] == 'Con Descuento']['Year'],
    y=comparativa[comparativa['flag_descuento'] == 'Con Descuento']['Total_Profit'],
    name='Beneficio con Descuento',
    mode='lines+markers',
    line=dict(color='darkblue', width=2),
))


fig.add_trace(go.Scatter(
    x=comparativa[comparativa['flag_descuento'] == 'Sin Descuento']['Year'],
    y=comparativa[comparativa['flag_descuento'] == 'Sin Descuento']['Total_Profit'],
    name='Beneficio sin Descuento',
    mode='lines+markers',
    line=dict(color='red', width=2),
))


fig.update_layout(
    barmode='group',
    title='Ventas y Beneficio por Año y Tipo de Descuento',
    xaxis_title='Año',
    yaxis_title='Cantidad',
    xaxis={'type': 'category'},
    legend_title_text='Tipo de Descuento y Métrica',
    template='plotly_white'
)

fig.show()
```

En este gráfico comprobamos que el beneficio de ventas con descuento es mucho menor que los beneficios que se obtienen con las ventas sin descuento.
Esto resalta y nos hace pensar que no es tan beneficioso tener mayor ventas con descuento, ya que el beneficiio es infinitamenente menor.
Deberíasmos transmitir a la empresa que quizas deban de cambiar su filosofia de ventas ya que no ocasiona rentabilidad y a la larga puede perjudicar.
Esto genera mas trabajo logistico y menos beneficio.


Conclusiones: con esto podemos justificar a la empresa que si continua con la dinamica de aplicar los descuentos tendrá un menor benefico
