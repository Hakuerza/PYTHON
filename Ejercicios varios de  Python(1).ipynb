```python
lista_a=[0,11,15,1,21,32,50,74,3,32]
print(lista_a)
```

```python
def multiplicacion_mayor_igual_50(lista_a):
  for i in lista_a:
    print(f"comprobando elemento {i} con multiplicacion {i*i}")
    if i*i>=50:
      print(f"elemento {i} es mayor o igual a 50")
      print("...")
    else:
      print(f"elemento (i) es menor a 50")
      print("...")
```

```python
multiplicacion_mayor_igual_50(lista_a)
```
